```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('processed_data.csv')
```

```python
# Extract relevant columns
facility_age = data['Facility Age']
facility_condition_numeric = data['Facility Condition Numeric']
fencing_condition_numeric = data['Fencing Available Numeric']
lighting_condition_numeric = data['Lighting Available Numeric']

# Create subplots for distributions
plt.figure(figsize=(14, 10))

# Facility Age Distribution
plt.subplot(2, 2, 1)
plt.hist(facility_age.dropna(), bins=20, edgecolor='black', alpha=0.7)
plt.title('Distribution of Facility Age', fontsize=14)
plt.xlabel('Age (years)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Facility Condition Numeric Distribution
plt.subplot(2, 2, 2)
plt.hist(facility_condition_numeric.dropna(), bins=10, edgecolor='black', alpha=0.7, color='orange')
plt.title('Distribution of Facility Condition (Numeric)', fontsize=14)
plt.xlabel('Condition Score', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Fencing Condition Numeric Distribution
plt.subplot(2, 2, 3)
plt.hist(fencing_condition_numeric.dropna(), bins=3, edgecolor='black', alpha=0.7, color='green')
plt.title('Distribution of Fencing Condition (Numeric)', fontsize=14)
plt.xlabel('Fencing Score', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Lighting Condition Numeric Distribution
plt.subplot(2, 2, 4)
plt.hist(lighting_condition_numeric.dropna(), bins=3, edgecolor='black', alpha=0.7, color='purple')
plt.title('Distribution of Lighting Condition (Numeric)', fontsize=14)
plt.xlabel('Lighting Score', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Set a style for the plots
sns.set(style="whitegrid")

# Facility Condition vs. Fencing Condition
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x="Facility Condition Numeric", hue="Fencing Available Numeric")
plt.title("Facility Condition vs. Fencing Condition")
plt.xlabel("Facility Condition Numeric")
plt.ylabel("Count")
plt.legend(title="Fencing Condition (1: Available, 0: Not Available)")
plt.show()

# Facility Condition vs. Lighting Condition
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x="Facility Condition Numeric", hue="Lighting Available Numeric")
plt.title("Facility Condition vs. Lighting Condition")
plt.xlabel("Facility Condition")
plt.ylabel("Count")
plt.legend(title="Lighting Condition (1: Available, 0: Not Available)")
plt.show()

# Facility Condition vs. Facility Age
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x="Facility Condition Numeric", y="Facility Age")
plt.title("Facility Condition vs. Facility Age")
plt.xlabel("Facility Condition")
plt.ylabel("Facility Age")
plt.show()
```
