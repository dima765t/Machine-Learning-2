```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = "Project_copy.xlsx"
sheet_name = "מתקנים לפי רשויות מעודכן 25.7.2"  # Adjust based on your dataset
data = pd.read_excel(file_path, sheet_name=sheet_name)

# Display the first few rows
data.head()
```

```python
# Create a dictionary mapping Hebrew column names to English
column_name_mapping = {
    "רשות מקומית": "Municipality",
    "ישוב": "Settlement",
    "מספר זיהוי": "ID",
    "סוג מתקן": "Facility Type",
    "שם המתקן": "Facility Name",
    "שכונה-רובע": "Neighborhood",
    "רחוב": "Street",
    "מספר בית": "House Number",
    "ציר X": "X Cord",
    "ציר Y": "Y Cord",
    "גוף מפעיל המתקן": "Operator",
    'בעלי המתקן': "Owner",
    "מספר המבנים": 'Building Amount',
    'טלפון איש קשר': "Contact Phone",
    'דואל איש קשר': "Contact eMail",
    'מספר מושבים': "Seat amount",
    "פנוי לפעילות": "Availability",
    "גידור קיים": "Fencing Available",
    "תאורה קיימת": "Lighting Available",
    "נגישות לנכים": "Accessibility",
    "חניה לרכבים": "Parking Available",
    "מצב המתקן": "Facility Condition",
    "מתקן תקני לתחרויות": "Competition Standard",
    "שימוש לתחרויות רשמיות": "Official Competition Use",
    "שנת הקמה": "Year Built",
    "משרת בית ספר": "School Served"
}

# Rename the columns in the dataset
data.rename(columns=column_name_mapping, inplace=True)

# Display the updated columns
print("Updated column names:")
print(data.columns)
```

```python
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
# Save the dataset with updated column names
data.to_csv("processed_data.csv", index=False)
```

```python
# Find and display unique values for each column
for column in data.columns:
    unique_values = data[column].unique()
    print(f"Column: {column}\nUnique Values: {unique_values}\n")
```

```python
availability_mapping = {
    "פתוח ללא הגבלה": "Very High",
    "מבוקר ועד לילה": "High",
    "בתיאום בלבד": "Low",
    "בוקר בלבד": "Medium"
}

data['Availability Category'] = data['Availability'].map(availability_mapping)
```

```python
fencing_mapping = {
    "כן": 1,
    "לא": 0
}

data['Fencing Available Numeric'] = data['Fencing Available'].map(fencing_mapping)
```

```python
lighting_mapping = {
    "כן": 1,
    "לא": 0
}

data['Lighting Available Numeric'] = data['Lighting Available'].map(lighting_mapping)
```

```python
condition_mapping = {
    "תקין ופעיל": 4,
    "פעיל וזקוק לשיפוץ": 3,
    "לא פעיל": 2,
    "לא תקין ולא פעיל": 1,
    "סגור לשימוש": 0
}

data['Facility Condition Numeric'] = data['Facility Condition'].map(condition_mapping)
```

```python
# Verify the newly created columns
print(data[['Availability', 'Availability Category', 
            'Fencing Available', 'Fencing Available Numeric', 
            'Lighting Available', 'Lighting Available Numeric', 
            'Facility Condition', 'Facility Condition Numeric']].head())
```

```python
current_year = 2025

# Calculate the age of the facility
data['Facility Age'] = current_year - data['Year Built']

# Ensure there are no negative or invalid ages
data['Facility Age'] = data['Facility Age'].apply(lambda x: x if x >= 0 else None)

# Display the first few rows to verify the new column
print(data[['Year Built', 'Facility Age']].head())
```

```python
# Save the dataset with updated column names
data.to_csv("processed_data.csv", index=False)
data.head()
```
